### Missing values in rating and size

```python
mean_rating = playstore['rating'].mean()
mean_size = playstore['size(kb)'].mean()
playstore['rating'] = playstore['rating'].fillna(mean_rating)
playstore['size(kb)'] = playstore['size(kb)'].fillna(mean_size)
```

```python
playstore.loc[playstore['type'].isnull()]
```

####  Convert 0 to 'Free' and 1 to 'Paid' in the type column. 
#### Convert the data type to string

```python
playstore['type']=playstore['type'].replace(0,'Free')
playstore['type']=playstore['type'].replace(1,'Paid')
playstore['type']=playstore['type'].astype(str)
print(playstore['type'].dtypes)
```

```python
playstore.dtypes
```

#### Convert rating from 'Object' data type to 'Integer'

```python
playstore['rating']=playstore['rating'].astype(int)
print(playstore['rating'].dtypes)
```

```python
playstore.loc[playstore['content_rating'].isnull()]
```

```python
playstore.dropna(how = 'any', inplace = True)
```

```python
playstore.isnull().sum()
```
